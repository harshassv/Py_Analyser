```python
def list_imports(lines):
    s_list=lines.split('\n')
    ans=[]
    o_ans=[]
    for i in s_list:
        if 'import' in i:
            i=i[i.index('import')+6:]
            if ',' not in i and 'as ' not in i:
                ans.append(i[i.index(' ')+1:])
            elif 'as ' not in i and ',' in i:
                ans.extend(i[i.index(' ')+1:].split(','))
            else:
                i=i.split(',')
                for j in i:
                    if 'as ' not in j:
                        ans.append(j)
                    else:
                        ans.append(j[j.index('as ')+3:])
    for i in ans:
        o_ans.append(i.strip()) #to trim white spaces
    return o_ans
```
